```python
# open raw data file :
with open("instagram_varied_data.txt", encoding='utf-8') as f:
    data = f.read()
```

```python
# defining a function to parse the data (parse means converting raw data into organised data)
def parse_chunk(chunk):
    chunk = chunk.strip() # removing blank spaces from the edges
    sep_chunk = chunk.split('\n') # splitting each chunk line as seen above chunks will give all the data of the chunks in a single line which we will separate it
    user_name = sep_chunk[0]
    no_of_posts = int(sep_chunk[1].split(" post")[0].replace(",",""))
    followers = sep_chunk[2].split(" follower")[0].replace(",","")
    no_of_followers = int(float(followers[:-1]) *1000) if followers.endswith("K") else int(float(followers[:-1]) *1000000) if followers.endswith("M") else int(followers)
    no_of_following = int(sep_chunk[3].split(" following")[0].replace(",",""))
    name =sep_chunk[4]
    if len(sep_chunk)>=6:
        type_of_page = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else:
        type_of_page = "unknown"
        bio = []
    # print(user_name,no_of_posts,no_of_followers,no_of_following,name,type_of_page,bio) just to check whether the function is running or not
    return{"user_name":user_name,"no_of_posts":no_of_posts,"no_of_followers":no_of_followers,"no_of_following":no_of_following,"name":name,
           "profession":type_of_page,"bio":bio}

parse_chunk(chunks[1])
```

```python
all_chunks=[]
for chunk in chunks:
    parsed_chunk=parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
    
print(len(all_chunks), all_chunks)
```
